```python
# Padronizando as variáveis numéricas.
scaler = StandardScaler()

independent_cols = df.drop('Rank', axis = 1).columns
numerical_data = df[independent_cols].copy()

transformed_x = scaler.fit_transform(numerical_data)
df.loc[:, independent_cols] = transformed_x
```

```python
# Definição de variáveis.
y_var = 'Rank'
x_var = independent_cols
```

### Verificando a multicolinearidade no modelo.

```python
# VIF - Multicolinearidade.
vif = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif_df = pd.DataFrame({'vif': vif[1:]}, index = x.columns.drop('const'))

def highlight_max(s):
  is_max = s > 10
  return ['background-color: orange' if v else '' for v in is_max]

vif_df.style.apply(highlight_max)
```

### Como todas as variáveis independentes possuem VIF <= 10, não há multicolinearidade. Então podemos considerar as 4 variáveis no modelo de regressão.

### Construindo o modelo de Regressão Linear para prever o Rank e descobrir quais atributos mais influem na variável resposta.

### Regressão Linear - Statsmodel

```python
Y = df[y_var]
x = df[x_var]
x = sm.add_constant(x)

model_sm = sm.OLS(Y,x).fit()
model_sm.summary()
```

### Verificando a independência dos resíduos com o teste de Durbin-Watson. Percebemos que o teste aponta evidência de correlação positiva. Ou seja, existe autocorrelação dos resíduos. Uma possibilidade, a fim de corrigir isso, seria a inserção de outras variáveis independentes importantes, ou até mesmo avaliar modelos alternativos.

### Verificando a Homecedasticidade dos Resíduos.

```python
# Gráfico de Resíduos versus Valores Previstos.
plt.figure(figsize=(8,5))
plt.scatter(y=model_sm.resid, x=model_sm.predict(x), color='silver', s=50, alpha=.6)
plt.hlines(y=0, xmin=-10, xmax=15, color='orange')
plt.ylabel('resíduo')
plt.xlabel('predict')
plt.show()
```

### O gráfico indica uma tendência ou padrão no espalhamento dos resíduos, logo uma não-homocedasticidade.

```python
#Teste de homocedasticidade
print('goldfeld quandt test:', round(sms.het_goldfeldquandt(model_sm.resid, 
                                                            model_sm.model.exog)[0],3))
print('p goldfeld quandt:', sms.het_goldfeldquandt(model_sm.resid, 
                                                   model_sm.model.exog)[1])
```

### Pelo teste de homocedasticidade, podemos perceber que o valor p é maior do que 0,05, indicando que não podemos rejeitar a hipótese nula de que os resíduos são homocedásticos.
